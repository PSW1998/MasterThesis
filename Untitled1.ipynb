```python
# Define the independent variables (X) and the dependent variable (y)
X = df_log[['BTC / Addresses, active, count', 'BTC / Transactions, transfers, value, mean, USD','NFCI','WM2NS']]
y = df_log['BTC / Capitalization, market, current supply, USD']

# Define the size of the rolling window
window_size = 52

# Loop through the rolling window to fit the regression model and make predictions
predictions = []
for i in range(window_size, len(df_log)):
    # Get the data in the current window
    X_window = X.loc[df_log.index[i-window_size+1]:df_log.index[i]]

    # Define the dependent variable in the current window
    y_window = y.loc[df_log.index[i-window_size+1]:df_log.index[i]]

    # Add a constant to the independent variables to include an intercept in the model
    X_window = sm.add_constant(X_window)

    # Fit the multivariate regression model using ordinary least squares (OLS)
    model = sm.OLS(y_window, X_window).fit()

    # Make a prediction for the next value in the time series
    next_X = [1] + list(X.iloc[i])
    next_prediction = model.predict(next_X)
    predictions.append(next_prediction[0])

# Add the predictions to the DataFrame
df_log['predictions'] = [None] * window_size + predictions

df_log[['predictions', 'BTC / Capitalization, market, current supply, USD']].plot()
```
